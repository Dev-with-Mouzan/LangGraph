#  What is LangGraph ? 
LangGraph is an open-source framework developed by LangChain for building reliable, stateful, and complex multi-actor AI agents. It extends LangChain to model agent workflows as graphs—nodes represent actions or LLM calls, and edges represent flow logic (loops, branching). This structure enables building agents that can reason, act, and maintain memory over long-running interactions.

# Key Features and Capabilities:
* <b>Graph-Based Architecture:</b> Defines AI workflows as graphs with nodes (computation steps) and edges (control flow).

* <b>Persistent State Management:</b> Built-in persistence allows agents to remember information across multiple turns,    enabling complex conversational AI and debugging.

* <b>Cyclical Flows & Loops:</b> Supports iterating through steps, which is crucial for agentic loops (e.g., plan, act, observe, re-plan).

* <b>Human-in-the-Loop:</b> Enables pausing the execution of a graph to allow for human review or intervention before continuing, which is crucial for high-stakes AI automation.

* <b>Orchestration:</b> Unlike simple, linear chains in LangChain, LangGraph provides fine-grained control over the decision-making flow and tool usage.
