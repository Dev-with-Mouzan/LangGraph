# What is Agentic AI?
An AI that can take up a task from a user and then work a user and then work toward completing it on it's own with minimal human guidance.

## Characteristics of Agentic AI Agents

- **Goal-driven behavior**
  - Operate with explicit objectives or mission statements.
  - Choose actions to achieve long-term outcomes, not just immediate responses.

- **Autonomy**
  - Make decisions without requiring constant human control.
  - Manage their own task selection, scheduling, and execution.

- **Planning and reasoning**
  - Form plans or action sequences to reach goals.
  - Evaluate possible strategies and update plans when conditions change.

- **Perception and environment awareness**
  - Sense and interpret relevant inputs from their environment.
  - Use observations to guide decisions and detect when goals are reached.

- **Memory and state**
  - Retain past information, decisions, and context across interactions.
  - Use history to improve future actions and avoid repeating mistakes.

- **Adaptability**
  - Adjust strategies based on feedback, new data, or changing goals.
  - Learn from experience or from external signals to improve performance.

- **Collaboration and communication**
  - Interact with humans, other agents, or systems as needed.
  - Share status, ask for help, and coordinate with peers.

- **Safety and constraint handling**
  - Respect rules, ethical constraints, and boundaries.
  - Monitor for unsafe actions and avoid harmful behavior.

- **Self-monitoring**
  - Assess their own performance and decide when to pause, retry, or escalate.
  - Detect uncertainty and request clarification when needed.

These traits distinguish agentic systems from simple reactive or prompt-based models by enabling purposeful, multi-step, context-aware action.
