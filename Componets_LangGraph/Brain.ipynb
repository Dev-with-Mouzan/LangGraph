# What is the Brain in an LLM Workflow?

The **Brain** is the **Large Language Model (LLM)** that powers the workflow. It is responsible for understanding the user's request, reasoning about it, making decisions, and determining the next step.

For example, the Brain can:
- Understand the user's query.
- Decide whether it can answer directly or needs a tool.
- Process the information.
- Generate the final response.

In **LangGraph**, the Brain is typically an **LLM node** that controls the flow of the workflow by making intelligent decisions.

## Example

**User:** *"What's the weather in Karachi today?"*

- The **Brain (LLM)** understands that the question requires real-time weather information.
- It decides to call a **Weather API**.
- After receiving the weather data, it generates a natural language response for the user.

**In simple terms:** The Brain is the decision-maker of an LLM workflow, while tools perform the actions requested by the Brain.
