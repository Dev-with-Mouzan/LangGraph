# What is Memory in LangGraph?

**Memory** in **LangGraph** is the ability of an application to **store and recall information** during or across conversations. It helps the AI remember previous interactions, workflow state, or important details so it can provide more relevant and consistent responses.

Memory allows the workflow to maintain context instead of treating every user request as completely new.

## Example

**User:** *"My name is Ali."*

Later, the user asks:

**User:** *"What's my name?"*

If memory is enabled, LangGraph remembers the previous information and responds:

**AI:** *"Your name is Ali."*

## Key Points

- Memory stores important information for later use.
- It helps maintain context across multiple steps or conversations.
- In LangGraph, memory is commonly used to save the workflow state and conversation history.
- Memory makes AI applications more personalized and context-aware.

# Types of Memory in LangGraph

LangGraph mainly supports **two types of memory**:

## 1. Short-Term Memory

Short-term memory stores information **during the current conversation or workflow**. It helps the AI remember previous messages and the current state while the workflow is running.

**Example:**

- **User:** "My favorite color is blue."
- **User:** "What is my favorite color?"

**AI:** "Your favorite color is blue."

> This memory is typically cleared when the conversation or workflow ends.

---

## 2. Long-Term Memory

Long-term memory stores information **across multiple conversations or sessions**. It allows the AI to remember user preferences or important details even after the conversation has ended.

**Example:**

- Today: **User:** "I'm a Python developer."
- Next week: **User:** "What programming language do I use?"

**AI:** "You use Python."

> This memory is stored in a database or external storage and can be retrieved in future conversations.

---

## Key Points

- **Short-Term Memory:** Stores information only for the current conversation or workflow.
- **Long-Term Memory:** Stores information across multiple conversations.
- Short-term memory is managed using the workflow state.
- Long-term memory typically uses external storage such as a database or vector store.
