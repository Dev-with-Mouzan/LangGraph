# What is an Orchestrator in LangGraph?

An **Orchestrator** in **LangGraph** is the component that **manages and controls the flow of the workflow**. It decides which node should execute next based on the current state and the workflow logic.

Instead of performing tasks itself, the orchestrator coordinates different nodes to complete the workflow efficiently.

## Example

Suppose a user asks:

**User:** *"Summarize the latest AI news."*

The orchestrator manages the workflow as follows:

1. Send the request to the **LLM node**.
2. The LLM decides that current news is needed.
3. Route the workflow to the **Search Tool**.
4. Send the search results back to the **LLM** for summarization.
5. Return the final summary to the user.

## Key Points

- The orchestrator controls the workflow execution.
- It determines which node runs next.
- It manages the flow between LLMs, tools, and other nodes.
- In LangGraph, the graph itself acts as the orchestrator by routing execution based on nodes, edges, and state.
