# What is a Supervisor in LangGraph?

A **Supervisor** in **LangGraph** is a special **LLM node** that manages multiple agents or workers. It analyzes the user's request, decides which agent should handle the task, and coordinates the overall workflow.

The supervisor does not perform every task itself. Instead, it delegates tasks to the most appropriate agent and combines their results when needed.

## Example

**User:** *"Research the latest AI trends and create a summary."*

The **Supervisor**:
1. Assigns the research task to a **Research Agent**.
2. Sends the collected information to a **Summarization Agent**.
3. Returns the final summary to the user.

## Key Points

- A Supervisor manages multiple agents.
- It decides which agent should perform a task.
- It coordinates the workflow and combines results.
- In LangGraph, a Supervisor is commonly used in **multi-agent workflows**.
