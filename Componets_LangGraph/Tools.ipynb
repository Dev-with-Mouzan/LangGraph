# What are Tools in LangGraph?

**Tools** in **LangGraph** are external functions or services that an LLM can use to perform tasks it cannot do on its own, such as fetching real-time data, performing calculations, querying databases, or calling APIs.

The LLM decides **when** to use a tool, and the tool executes the requested action and returns the result.

## Example

**User:** *"What is the weather in Lahore today?"*

- The **LLM** determines that real-time weather information is required.
- It calls a **Weather Tool (API)**.
- The tool returns the current weather data.
- The LLM generates a user-friendly response.

## Key Points

- Tools extend the capabilities of the LLM.
- They provide access to real-time data and external systems.
- The LLM decides when a tool should be used.
- Common tools include search APIs, calculators, databases, and custom functions.
