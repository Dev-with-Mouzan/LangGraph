# What is Conditional WorkFlow in LangGraph?


 A conditional workflow in LangGraph is a directed graph where execution paths split based on evaluated conditions, choosing different nodes/transitions depending on runtime values.



# Key Component
 **Node**: step or action; **Condition**: boolean/expression evaluated at a node; **Branch/Edge**: outgoing path selected when a condition holds; **State**: data used by conditions.
- **How it works**: At a condition node, LangGraph evaluates expressions against the current state (variables, model outputs). The true branch is followed (or the first matching branch), updating state and continuing until an end node.
- **When to use**: routing user intents, multi-step forms, validation flows, A/B paths, error handling, or any logic that depends on runtime data.
- **Minimal example**: If `sentiment > 0` → `PositiveHandler` else → `NegativeHandler`.
