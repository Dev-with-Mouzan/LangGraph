```python
from langchain_ollama import ChatOllama
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
class input_data(TypedDict):
    topic: str
    outline: str
    details: str
```

```python
def create_outline(input:input_data) -> input_data:
    llm = ChatOllama(model="tinyllama")
    prompt = f"Create a detailed outline for an article on the topic: {input['topic']}"
    response = llm.invoke(prompt)
    input['outline'] = response.content
    return input
```

```python
def create_details(input:input_data) -> input_data:
    llm = ChatOllama(model="tinyllama")
    prompt = f"Expand the following outline\n{input['outline']} into a detailed article"
    response = llm.invoke(prompt)
    input['details'] = response.content
    return input
```

```python
graph=StateGraph(input_data)
graph.add_node("outline_creator",create_outline)
graph.add_node("details_creator",create_details)
graph.add_edge(START,"outline_creator")
graph.add_edge("outline_creator","details_creator")
graph.add_edge("details_creator",END)
workflow=graph.compile()
```

```python
finall_result=workflow.invoke({"topic":"The impact of AI on modern education"})
```
