# What is Sequential Workflow?

A Sequential Workflow in LangGraph is the simplest workflow pattern in which tasks are executed one after another in a predefined order. Each node begins execution only after the previous node has completed successfully. The shared state is updated at every step and passed along the workflow until the graph reaches the END node.

# Characteristics
* Linear execution: Nodes execute in a fixed sequence.
* Deterministic behavior: The execution path remains the same for every run unless the graph definition changes.
* Shared state: All nodes operate on a common state object, allowing information produced by earlier nodes to be used by later ones.
* Single execution path: There is only one route from the START node to the END node.
* Dependency between steps: Each step typically depends on the output of the previous step.

# How it Work
A sequential workflow starts at the START node and follows a linear path through a series of connected nodes. Each node:

<h4>Step 1</h4>
Receives the current shared state.
<h4>Step 2</h4>
Performs a specific operation.
<h4>Step 3 </h4>
Updates the state.
<h4>Step 4 </h4>
Passes the updated state to the next node.

This process continues until the workflow reaches the END node, where execution stops.
