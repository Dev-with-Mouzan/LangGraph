```python
from langgraph.graph import START,END,StateGraph
from langchain_ollama import ChatOllama
from typing import TypedDict
```

```python
class Resume_ana(TypedDict):
    resume_text:str
    summary:str
    Keyword:str
    Ats_score:str
```

```python
llm =ChatOllama(model="llama3.2:1b ")
```

```python
def summarizer(state:Resume_ana)->Resume_ana:
    prompt = f"""
    You are a resume analyzer. Your task is to analyze the resume text and provide a summary
    
    Resume Text:
    {state['resume_text']}"""
    
    
    response = llm.invoke(prompt).content
    

    
    return {
        
        "summary": response
    }
```

```python
def keyword_extractor(state:Resume_ana)->Resume_ana:
    prompt = f"""
    You are a resume analyzer. Your task is to extract relevant keywords from the resume text.
    
    Resume Text:
    {state['resume_text']}"""
    
    
    response = llm.invoke(prompt).content
    
    return {
        
        "Keyword": response
    }
```

```python
def Ats_score_extractor(state:Resume_ana)->Resume_ana:
    prompt = f"""
    You are a resume analyzer. Your task is to provide an ATS score for the resume text out of 100 points.
    
    Resume Text:
    {state['resume_text']}"""
    
    
    response = llm.invoke(prompt).content
    
    return {
        
        "Ats_score": response
    }
```

```python
graph=StateGraph(Resume_ana)
```

```python
graph.add_node("summarizer",summarizer)
graph.add_node("keyword_extractor",keyword_extractor)
graph.add_node("Ats_score_extractor",Ats_score_extractor)

graph.add_edge(START,"summarizer")
graph.add_edge(START,"keyword_extractor")
graph.add_edge(START,"Ats_score_extractor")

graph.add_edge("summarizer",END)
graph.add_edge("keyword_extractor",END)
graph.add_edge("Ats_score_extractor",END)
```

```python
app=graph.compile()
```

```python
app.invoke({"resume_text":"I am a software engineer with 5 years of experience in developing web applications using Python and JavaScript. I have a strong background in data structures and algorithms, and I am proficient in frameworks such as Django and React. I have also worked on cloud platforms like AWS and Azure, and I have experience with CI/CD pipelines and containerization using Docker."})
```
