# What is Thread_Id ?


A **Thread ID** is a **unique identifier** that represents a single execution thread or conversation in a LangGraph application.

When using **LangGraph Persistence**, the Thread ID tells LangGraph:

> **"Load and save the state for this specific workflow or conversation."**

Think of it as a **unique key** that allows LangGraph to distinguish one workflow execution from another.

---

# Why Do We Need a Thread ID?

Without persistence, every time you invoke a graph, it starts with a fresh state.

For example:

```text
User:
My name is John.

↓

Graph stores:
name = "John"

↓

Execution Ends
```

If the user sends another message:

```text
What's my name?
```

Without persistence:

```text
Graph starts from scratch.

State = {}
```

The graph has forgotten everything.

---

# The Solution: Persistence

With persistence enabled, LangGraph saves the workflow state.

However, LangGraph needs to know **which saved state** belongs to which user or workflow.

This is where the **Thread ID** comes in.

---

# What Does Thread ID Do?

A Thread ID acts as a **storage key**.

For example:

```text
Thread ID = "user_101"

↓

State

{
    "messages": [...],
    "name": "John"
}
```

Later, when the same Thread ID is used:

```text
Thread ID = "user_101"

↓

LangGraph loads

{
    "messages": [...],
    "name": "John"
}
```

The conversation continues from where it left off.

---

# Simple Analogy

Imagine a library.

Every visitor receives a locker.

```text
Alice → Locker 101

Bob → Locker 102

Charlie → Locker 103
```

Each person's belongings are stored in their own locker.

When Alice returns, she uses:

```text
Locker 101
```

to retrieve her items.

The **Thread ID** works exactly like the locker number.

It tells LangGraph which stored state belongs to the current execution.

---

# Without Thread ID

Suppose two users are chatting with the same AI.

### User A

```text
Hi

My name is Alice.
```

### User B

```text
Hello

My name is Bob.
```

If no Thread ID exists:

```text
Shared State

↓

Name = ?
```

LangGraph cannot determine whether the state belongs to Alice or Bob.

This could result in mixed or overwritten conversations.

---

# With Thread ID

Each user gets a unique Thread ID.

```text
Thread ID

alice_001

↓

State

{
    name: Alice
}
```

```text
Thread ID

bob_001

↓

State

{
    name: Bob
}
```

Now each conversation is isolated.

---

# Visual Representation

```text
            Users
              │
     ┌────────┴────────┐
     ▼                 ▼
 Thread ID A      Thread ID B
     │                 │
     ▼                 ▼
 Saved State A    Saved State B
```

Every Thread ID points to its own persisted state.

# Example Conversation

### First Request

```text
Thread ID = user_123

User:
My favorite language is Python.
```

LangGraph saves:

```python
{
    "favorite_language": "Python"
}
```

---

### Second Request

```text
Thread ID = user_123

User:
What's my favorite language?
```

LangGraph loads:

```python
{
    "favorite_language": "Python"
}
```

Response:

```text
Your favorite language is Python.
```

---

### Different Thread

```text
Thread ID = user_456

User:
What's my favorite language?
```

LangGraph loads:

```python
{}
```

Response:

```text
I don't know your favorite language yet.
```

The state is different because the Thread ID is different.

# Where Is Thread ID Used?

Thread IDs are commonly used in:

- Chatbots
- AI assistants
- Customer support systems
- Multi-user applications
- Agent workflows
- Long-running processes
- Human-in-the-loop workflows
- Resumable graph executions

---

# How Thread ID Works with Persistence

```text
User Request
      │
      ▼
Thread ID
      │
      ▼
Load Saved State
      │
      ▼
Execute Graph
      │
      ▼
Update State
      │
      ▼
Save State Using Same Thread ID
```

This allows the graph to resume from its previous state on the next invocation.

---

# Common Sources of Thread IDs

A Thread ID can be generated from various identifiers, depending on your application:

- User ID
- Session ID
- Chat ID
- Conversation ID
- Order ID
- Workflow ID
- A generated UUID

The important requirement is that it is **unique for each independent workflow or conversation**.

---

# Benefits of Using Thread ID

- Keeps conversations separate.
- Enables memory across multiple requests.
- Supports resumable workflows.
- Prevents state conflicts between users.
- Makes long-running AI agents possible.
- Allows persistence across application restarts.

---

# Best Practices

- Use a unique Thread ID for each conversation or workflow.
- Reuse the same Thread ID when you want to continue an existing conversation.
- Generate Thread IDs using reliable identifiers such as user IDs or UUIDs.
- Avoid sharing the same Thread ID across unrelated users or workflows.
- Store the Thread ID securely if the application needs to resume executions later.

---

# Thread ID vs Session ID

| Thread ID | Session ID |
|-----------|------------|
| Identifies a persisted workflow or conversation | Identifies a user's current application session |
| Used by LangGraph to load and save state | Used by web applications to manage user sessions |
| Can persist across multiple sessions | Often expires when the session ends |
| Controls graph persistence | Controls application authentication and session management |

> **Note:** While a Session ID can sometimes be used as a Thread ID, they are not the same concept. A Thread ID is specifically used by LangGraph to identify and retrieve the correct persisted workflow state.

---

# Summary

A **Thread ID** is a unique identifier that links a LangGraph execution to its persisted state. Whenever a graph is invoked with persistence enabled, LangGraph uses the Thread ID to load the previous state, execute the workflow, and save the updated state back under the same identifier. This makes it possible to build chatbots, AI agents, and long-running workflows that remember past interactions while keeping each user's data isolated.
