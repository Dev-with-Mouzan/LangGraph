```python
from langgraph.checkpoint.sqlite import SqliteSaver
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Annotated
from langchain_ollama import ChatOllama
from operator import add
from langchain_core.messages import HumanMessage, AIMessage
```

```python
class memory(TypedDict):
    chat: Annotated[list, add]
```

```python
llm=ChatOllama(model="llama3.2:1b")
```

```python
def chat_node(state:memory)->memory:
    message=state["chat"]
    repsonse=llm.invoke(message).content
    return {"chat":[AIMessage(content=repsonse)]}
```

```python
graph=StateGraph(memory)
```

```python
graph.add_node("Chat_Bot",chat_node)

graph.add_edge(START,"Chat_Bot")
graph.add_edge("Chat_Bot",END)
```

```python
with SqliteSaver.from_conn_string("cha.db") as connection_db:

    workflow = graph.compile(
        checkpointer=connection_db
    )


    result = workflow.invoke(
        {
            "chat":[HumanMessage(content="i am ali")]
        },
        config={
            "configurable":{
                "thread_id":"user1"
            }
        }
    )


print(result)
```
